```python
import glucifer
```

```python
saved = glucifer.Viewer('subduction')
```

```python
#Re-visualise the final timestep
#saved.step = saved.steps[-1]
for step in saved.steps:
    for name in saved:
        print(name)
        fig = saved[name]
        fig.quality = 2
        fig.properties["title"] = "Timestep " + str(step)
        fig.show()
        fig.save_image(name + str(step) )
```
