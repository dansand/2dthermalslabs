```python
#rdir = '/workspace'
rdir = '../'
dat_files0 = natsort.natsort(glob.glob(rdir + 'temp_results/files/*metricSwarm*'))
dat_files1 = natsort.natsort(glob.glob(rdir + 'temp_results/files/*interfaceStress*'))
dat_files0
```

```python
data_list0 = []
for f in dat_files0[1:]:
    with h5py.File(f,'r') as hf:
        #print('List of arrays in this file: \n', hf.keys())
        data = hf.get('data')
        np_data = np.array(data)
        data_list0.append(np_data)
```

```python
data_list1 = []
for f in dat_files1[1:]:
    with h5py.File(f,'r') as hf:
        print('List of arrays in this file: \n', hf.keys())
        data = hf.get('data')
        np_data = np.array(data)
        data_list1.append(np_data)
```
